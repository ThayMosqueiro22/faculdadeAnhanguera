```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

df = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.barplot(hue="time", y="total_bill", data=df, estimator=sum, errorbar=None, palette="Set2")
plt.xlabel('Periodo(Time)')
plt.ylabel('Total de Gastos')
plt.title('Total de Gastos por Perodo (Almoo ou Jantar)')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='total_bill', data=df)
plt.xlabel('Periodo(Time)')
plt.ylabel('Mdia de Gastos')
plt.title('Mdia de Gastos por Perodo (Almoo ou Jantar)')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(hue='time', y='total_bill', data=df, palette="Set3")
plt.xlabel('Perodo (Time)')
plt.ylabel('Mdia da Gorjeta')
plt.title('Mdia da Gorjeta por Perodo(Almoo ou Jantar)')
plt.show()
```
