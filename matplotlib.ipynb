```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,1,3,5]

plt.plot(x,y)

plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Exemplo Gráfico de Linhas")

plt.show()
```
