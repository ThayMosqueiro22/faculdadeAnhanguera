```python
import matplotlib.pyplot as plt

class Livro:
  def __init__(self, titulo, autor, ano_publicacao):
    self.titulo = titulo
    self.autor = autor
    self.ano_publicacao = ano_publicacao

  def __str__(self):
    return f" {self.titulo} por {self.autor}, Publicação em {self.ano_publicacao}"

biblioteca = []

anos = []

def add_livro (titulo, autor, ano_publicacao):
  novo_livro = Livro(titulo, autor, ano_publicacao)
  biblioteca.append(novo_livro)
  anos.append(ano_publicacao)
  print(f"O livro '{titulo}' foi adicionado à biblioteca.")

def listar_livros():
  print("Livros na biblioteca:")
  for livro in biblioteca:
    print(livro)

add_livro("Dom Quixote","Meguel de Cervantes",1605)
add_livro("O Pequeno Príncipe", "Antoine de Saint-Exupéry", 1943)
add_livro("1984", "George Orwell", 1949)
add_livro("A Revolução dos Bichos", "George Orwell", 1945)
add_livro("O Hobbit", "J.R.R. Tolkien", 1937)


listar_livros()

anos = list(set(anos))
anos.sort()

contagem_por_ano = [anos.count(ano_atual) for ano_atual in anos]

plt.plot(anos,contagem_por_ano, marker='o')
plt.xlabel('Ano')
plt.ylabel('Número de Livros')
plt.title('Distribuição de livros na Biblioteca por Ano de Puplicação')

for i, valor in enumerate(contagem_por_ano):
  plt.text(anos[i], valor, str(valor), ha='center', va='bottom')

plt.grid(True)
plt.show()
```
